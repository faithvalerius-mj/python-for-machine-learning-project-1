```python
#is this effecient? no

#for instance, let's assume that we would liek to show how to programmatically define a list of elements that
#satisfy a very complet criteria that cannot be define directly using the filter operations

#a simplet example in our case will be displaying the row where the number of proteins is greater than 2
#and the ratio between the rating and vitamins is greater than 2
```

```python
#We define a boolean list with no values to start with
items_to_display = []

#we start by itterating through the list of elements of the dataframe
for index, row in df.iterrows():
    #for each element we will take store the value of protein, rating, and vitamins in their own variables
    protein = row["protein"]
    vitamins = row["vitamins"]
    rating = row["rating"]

    #then we calculate the ratio rating/vitamins
    if vitamins > 0:
        ratio = rating/vitamins
    else:
        ratio = 0

    #now we check the value of the ratio and number of protein. If the conditions are satisfied, we display the row by setting the True
    #or false in theassociated boolean list we will later pass
    if protein > 2 and ratio >2:
        items_to_display.append(True)
    else:
        items_to_display.append(False)

#when the loop itteration is over, we display the result of our boolean list
print(items_to_display)

#and we pass the boolean list in the [] operation to show the rows where the data satisfies our define conditions
df[items_to_display]
```
